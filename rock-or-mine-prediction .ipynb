## importing required modules

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loding dataset
df= pd.read_csv("sonar data.csv", header=None)
```

```python
df.head()
```

```python
# checking number of rows and columns
df.shape
```

```python
# checking unique values in target column
df[60].value_counts()
```

```python
df[60]
```

```python
# splitting input and target columns
x= df.drop(columns=[60])
y=df[60]
```

```python
# splitting data into training and testing data
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
# training model using LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
# checking accuracy of tested data
accuracy_score(y_test, y_pred)
```

```python
# checking accuracy of training data
x_train_pred= lr.predict(x_train)
accuracy_score(x_train_pred, y_train)
```

```python
# checking accuracy using cross_val_score
np.mean(cross_val_score(lr, x, y, scoring= "accuracy", cv=29))
```
